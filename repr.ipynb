```python
import matplotlib.pyplot as plt
import networkx as nx


G = nx.DiGraph()

nodes = ["Модуль сбора данных", "База данных", "Модуль аналитики", 
         "Управление ценообразованием", "Пользовательский интерфейс", "Система безопасности",
         "Решения для маркетплейсов"]
G.add_nodes_from(nodes)

edges = [("Модуль сбора данных", "База данных"), 
         ("База данных", "Модуль аналитики"), 
         ("Модуль аналитики", "Управление ценообразованием"), 
         ("Управление ценообразованием", "База данных"), 
         ("Пользовательский интерфейс", "База данных"), 
         ("Пользовательский интерфейс", "Управление ценообразованием"),
         ("Система безопасности", "База данных"),
         ("Решения для маркетплейсов", "Управление ценообразованием")]
G.add_edges_from(edges)

plt.figure(figsize=(15, 10))
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=4000, 
        edge_color="gray", linewidths=1, font_size=12, arrowsize=20)

edge_labels = {("Модуль сбора данных", "База данных"): "Поток данных",
               ("База данных", "Модуль аналитики"): "Доступ к данным",
               ("Модуль аналитики", "Управление ценообразованием"): "Анализ и прогнозы",
               ("Управление ценообразованием", "База данных"): "Обновление цен",
               ("Пользовательский интерфейс", "База данных"): "Настройки и аналитика",
               ("Пользовательский интерфейс", "Управление ценообразованием"): "Управление и обратная связь",
               ("Система безопасности", "База данных"): "Защита данных",
               ("Решения для маркетплейсов", "Управление ценообразованием"): "Специализированные решения"}

nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='darkred')
plt.title("Архитектура сервиса 'Репрайсер'", fontsize=16)
plt.show()
```
